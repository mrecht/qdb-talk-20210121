```julia
using LibPQ
```

```julia
using Tables
```

```julia
using Plots
```

```julia
plotly()
```

```julia
conn = LibPQ.Connection("dbname=qdb user=admin password=quest port=8812")
```

```julia
r1 = execute(conn, """select ts, (max(max_price)-min(min_price))/min(min_price) as iv from xetra where mnemonic='WDI' sample by 1M""")
```

```julia
data1 = columntable(r1)
```

```julia
plot(data1.ts, data1.iv, title="1 Month")
```

```julia
r1 = execute(conn, """select ts, (max(max_price)-min(min_price))/min(min_price) as iv from xetra where mnemonic='WDI' and  ts >'2020-04-01' and ts<'2020-08-01' sample by 7d""")
```

```julia
data1 = columntable(r1)
```

```julia
plot(data1.ts, data1.iv, title="7d, between 2020-04-01 to 2020-08-01")
```

```julia
r1 = execute(conn, """select ts, (max(max_price)-min(min_price))/min(min_price) as iv from xetra where mnemonic='WDI' and ts >'2020-06-01' and ts<'2020-07-01' sample by 7d""")
```

```julia
data1 = columntable(r1)
```

```julia
plot(data1.ts, data1.iv, title="1d, between 2020-06-01 and 2020-07-01")
```

```julia
r1 = execute(conn, """select ts, (max(max_price)-min(min_price))/min(min_price) as iv from xetra where mnemonic='WDI' and  to_str(ts,'yyyy-MM-dd') = '2020-06-25' sample by 15m""")
```

```julia
data1 = columntable(r1)
```

```julia
plot(data1.ts, data1.iv, title="15m, 2020-06-25")
```

```julia
r1 = execute(conn, """select ts, (max(max_price)-min(min_price))/min(min_price) as iv, sum(traded_volume)/10000000 as v from xetra where mnemonic='WDI' and ts >'2020-06-10' and ts<'2020-07-01' sample by 5m""")
```

```julia
data1 = columntable(r1)
```

```julia
plot(data1.ts, [data1.iv data1.v], label=["iv" "v (scaled by 10000000)"], title="5m, between 2020-06-10 and 2020-07-01")
```
